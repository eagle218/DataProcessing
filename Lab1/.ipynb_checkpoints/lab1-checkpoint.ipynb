```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Function to read specific column from CSV file
def read_column_from_csv(filename, column_name):
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        return [int(row[column_name]) for row in reader if row[column_name].isdigit()]

# File path and column name
filename = 'russia_losses_equipment.csv'
column_name = "APC"  

# Step 2: Extracting APC losses
apc_losses = read_column_from_csv(filename, column_name)

# Step 3: Analyzing BTR losses

# Convert the data into a Numpy array for easier manipulation
apc_array = np.array(apc_losses)

# 3.1 Daily losses (assuming apc_array already represents daily losses)

# 3.2 Find the five largest daily losses
top_5_losses = np.sort(apc_array)[-5:][::-1]  # Sort and take the top 5 values in descending order

# 3.3 Total losses in 2023 (assuming last entry corresponds to today)
# For simplicity, we'll take the last 365 entries
losses_2023 = np.sum(apc_array[-365:])

# 3.4 Average losses over the last 100-500 days (slice the data accordingly)
average_losses_100_500_days = np.mean(apc_array[-500:-100])

# Step 4: Plotting the data

# Extract the last 200 days for plotting
last_200_days = apc_array[-200:]

# Plotting
plt.figure(figsize=(8, 6), dpi=100)
plt.plot(range(len(last_200_days)), last_200_days, linestyle='--', color='b', label='BTR Losses')

# Customizing the plot
plt.title("BTR Losses over the Last 200 Days")
plt.xlabel("Days")
plt.ylabel("Losses")
plt.legend()
plt.grid(True)

# Save the plot as a PNG file
plt.savefig("btr_losses_last_200_days.png", format="png")

# Show the plot
plt.show()

# Print results for verification
print("Top 5 Daily Losses:", top_5_losses)
print("Total BTR Losses in 2023:", losses_2023)
print("Average Losses (100-500 days):", average_losses_100_500_days)
```
