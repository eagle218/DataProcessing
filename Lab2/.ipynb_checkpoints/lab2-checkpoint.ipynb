```python
import pandas as pd
# 4
# Replace 'your_file.csv' with the actual file path
data = pd.read_csv('NationalNames.csv')

# Перегляд перших рядків даних:

data.head(8)
```

```python
# Перегляд останніх рядків даних:

data.tail(8)
```

```python
# Список усіх колонок у датасеті:
data.columns
```

```python
# Підрахунок кількості унікальних імен:

unique_names_count = data['Name'].nunique()
print(unique_names_count)
```

```python
# Кількість унікальних імен за кожен рік:

unique_names_per_year = data.groupby('Year')['Name'].nunique()
unique_names_per_year.head() 

```

```python
# Рік із максимальною кількістю унікальних імен:

year_with_max_unique_names = unique_names_per_year.idxmax()
max_unique_names = unique_names_per_year.max()
print(year_with_max_unique_names, max_unique_names)
```

```python
# Найпопулярніше ім’я у 2008 році:

most_popular_name_2008 = data[data['Year'] == 2008].sort_values(by='Count', ascending=False).iloc[0]
print(most_popular_name_2008['Name'])
```

```python
# # Аналіз імені "Jacob" у жіночій статі:

jacob_female = data[(data['Name'] == 'Jacob') & (data['Gender'] == 'F')]
jacob_most_popular_year = jacob_female[jacob_female['Count'] == jacob_female['Count'].max()]
jacob_most_popular_year
```

```python
# Пошук гендерно-нейтральних імен:

gender_neutral_counts = data.groupby(['Year', 'Name'])['Gender'].nunique().reset_index()
gender_neutral_counts = gender_neutral_counts[gender_neutral_counts['Gender'] == 2]
gender_neutral_per_year = gender_neutral_counts.groupby('Year').size()
year_with_most_gender_neutral_names = gender_neutral_per_year.idxmax()
most_gender_neutral_names = gender_neutral_per_year.max()
print(year_with_most_gender_neutral_names, most_gender_neutral_names)
```

```python
# Загальна кількість народжень за кожен рік:

total_births_per_year = data.groupby('Year')['Count'].sum()
total_births_per_year.head()  # Display the first 5 rows
```

```python
# Рік із найбільшою кількістю народжень:

year_with_most_births = total_births_per_year.idxmax()
most_births = total_births_per_year.max()
print(year_with_most_births, most_births)
```

```python
# Кількість народжень за гендером у різні роки:

births_by_gender_per_year = data.pivot_table(values='Count', index='Year', columns='Gender', aggfunc='sum')
births_by_gender_per_year.head()  # Display the first 5 rows
```

```python
# Роки, коли народилось більше дівчат:
years_with_more_girls = (births_by_gender_per_year['F'] > births_by_gender_per_year['M']).sum()
print(years_with_more_girls)
```

```python
# Кількість хлопчиків із ім’ям "Barbara":
barbara_boys_count = data[(data['Name'] == 'Barbara') & (data['Gender'] == 'M')]['Count'].sum()
print(barbara_boys_count)
```

```python
# Кількість років у спостереженні:

years_observed = data['Year'].nunique()
print(f'Спостереження проводилось {years_observed} років')
```

```python
# Популярні гендерно-нейтральні імена за всі роки:
gender_neutral_names = data.groupby('Name')['Year'].nunique()
popular_gender_neutral_names = gender_neutral_names[gender_neutral_names == data['Year'].nunique()].index.tolist()
popular_gender_neutral_names
```

```python
# Найпопулярніше з найменш популярних імен:

name_counts = data.groupby('Name')['Count'].sum()
least_popular_name = name_counts[name_counts == name_counts.min()].index[0]
print(f'Наиболее популярное из непопулярных имен – это {least_popular_name}. Им называли {name_counts.min()} раз')
```

```python
# Графік для імен "John" (жінки) та "Mary" (чоловіки):
import matplotlib.pyplot as plt

john_female = data[(data['Name'] == 'John') & (data['Gender'] == 'F')]
mary_male = data[(data['Name'] == 'Mary') & (data['Gender'] == 'M')]

plt.plot(john_female['Year'], john_female['Count'], label='John (Female)')
plt.plot(mary_male['Year'], mary_male['Count'], label='Mary (Male)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Female Name "John" and Male Name "Mary" Over the Years')
plt.legend()
plt.show()
```
