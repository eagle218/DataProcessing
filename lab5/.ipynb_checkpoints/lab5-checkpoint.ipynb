```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt

# Определяем словарь цветов с необходимыми цветами
color_dict = {
    'red': (0, 0, 255),
    'blue': (255, 0, 0),
    'cian': (255, 255, 0),  # "cian" в данных, соответствует "cyan"
    'magenta': (255, 0, 255),
    'yellow': (0, 255, 255),
}

# Шаг 1: Загрузка данных из Excel-файла
df = pd.read_excel('DataSet/kr1.xlsx')

# Устанавливаем индекс для выбора строки с нужным изображением
index = 4  # Измените по необходимости (индексация с 0)
data_row = df.iloc[index]

# Извлекаем параметры из строки данных
input_image_name = data_row['file name']

# Извлекаем 'line width' из Excel-файла
line_width = int(data_row['line width'])

# Устанавливаем длину дужек очков (фиксированное значение)
temple_length = 60  # значение по умолчанию

# Обрабатываем 'image size', если нужно
if 'image size' in data_row and not pd.isna(data_row['image size']):
    # Извлекаем ширину и высоту из формата '300x300'
    size_str = str(data_row['image size'])
    width_str, height_str = size_str.lower().split('x')
    width, height = int(width_str), int(height_str)
else:
    width, height = None, None  # или установите значения по умолчанию

# Извлекаем 'glasses color' и преобразуем в BGR
if 'glasses color' in data_row and not pd.isna(data_row['glasses color']):
    color_name = data_row['glasses color'].lower()
    glasses_color = color_dict.get(color_name)
    if glasses_color is None:
        
        glasses_color = color_dict['yellow']  # Цвет по умолчанию
else:
    glasses_color = color_dict['yellow']  # Цвет по умолчанию

# Шаг 3: Определяем функцию для добавления очков с увеличенными линзами и корректной перемычкой
def add_glasses_to_face(image_path, output_path, line_width, temple_length=60, resize_dims=None, glasses_color=(0, 255, 255)):
    # Загружаем изображение
    image = cv2.imread(image_path)

    # Проверяем успешность загрузки изображения
    if image is None:
        print(f"Ошибка: Изображение '{image_path}' не найдено.")
        return

    # Изменяем размер изображения, если заданы размеры
    if resize_dims is not None:
        image = cv2.resize(image, resize_dims)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Загрузка классификаторов Хаара для распознавания лица и глаз
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    # Используем 'haarcascade_eye.xml' для распознавания глаз
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
    
    # Обнаружение лиц
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
   

    if len(faces) == 0:
        
        cv2.imwrite(output_path, image)  # Сохраняем изображение, даже если лица не найдены
        return

    for (x, y, w, h) in faces:
        # Определяем область интереса для распознавания глаз
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = image[y:y+h, x:x+w]
        
        # Обнаружение глаз внутри области лица
        eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=3, minSize=(10, 10))
       

        if len(eyes) < 2:
            
            # Попробуем изменить параметры детекции глаз
            eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.05, minNeighbors=3, minSize=(5, 5))
            
            if len(eyes) < 2:
                continue  # Переходим к следующему лицу

        # Сортируем глаза по размеру (площади) и выбираем два самых больших
        eyes = sorted(eyes, key=lambda ex: ex[2]*ex[3], reverse=True)[:2]

        if len(eyes) < 2:
            continue

        # Сортируем глаза по координате x, чтобы выбрать левый и правый глаз
        eyes = sorted(eyes, key=lambda ex: ex[0])

        (ex1, ey1, ew1, eh1), (ex2, ey2, ew2, eh2) = eyes[:2]

        # Вычисляем центры и радиус для кругов
        eye1_center = (x + ex1 + ew1 // 2, y + ey1 + eh1 // 2)
        eye2_center = (x + ex2 + ew2 // 2, y + ey2 + eh2 // 2)
        # Увеличиваем коэффициент для радиуса, чтобы сделать линзы и оправу больше
        radius = int(0.5 * (ew1 + ew2) / 2)

        # Сдвигаем центры линз немного наружу для корректного расположения оправы
        eye1_center = (eye1_center[0] - int(0.1 * radius), eye1_center[1])
        eye2_center = (eye2_center[0] + int(0.1 * radius), eye2_center[1])

        # Рисуем круги вокруг глаз для очков (линзы и оправа)
        cv2.circle(image, eye1_center, radius, glasses_color, line_width)
        cv2.circle(image, eye2_center, radius, glasses_color, line_width)
        
        # Корректируем положение перемычки
        bridge_y = int((eye1_center[1] + eye2_center[1]) / 2 - 0.2 * radius)
        # Рисуем перемычку между кругами
        cv2.line(image, (eye1_center[0] + int(0.7 * radius), bridge_y), (eye2_center[0] - int(0.7 * radius), bridge_y), glasses_color, line_width)
        
        # Рисуем дужки очков (слегка поднимаем их вверх)
        cv2.line(image, (eye1_center[0] - radius, eye1_center[1]),
                 (eye1_center[0] - radius - temple_length, eye1_center[1] - int(0.2 * radius)), glasses_color, line_width)
        cv2.line(image, (eye2_center[0] + radius, eye2_center[1]),
                 (eye2_center[0] + radius + temple_length, eye2_center[1] - int(0.2 * radius)), glasses_color, line_width)
        break  # Обрабатываем только первое обнаруженное лицо

    # Сохраняем изображение с эффектом очков
    cv2.imwrite(output_path, image)
    
input_image_path = 'Images/' + input_image_name
output_image_path = 'output_' + input_image_name

# Подготавливаем размеры для изменения изображения, если они заданы
resize_dims = (width, height) if width and height else None

# Вызываем функцию, передавая line_width из Excel-файла
add_glasses_to_face(input_image_path, output_image_path, line_width, temple_length, resize_dims, glasses_color)

# Загружаем изображения для отображения
input_image = cv2.imread(input_image_path)

# Проверяем успешность загрузки входного изображения
if input_image is None:
    print(f"Ошибка: Входное изображение '{input_image_path}' не найдено.")
else:
    # Изменяем размер входного изображения для консистентности отображения
    if resize_dims is not None:
        input_image = cv2.resize(input_image, resize_dims)

    output_image = cv2.imread(output_image_path)

    # Преобразуем изображения из BGR в RGB для отображения в matplotlib
    input_image_rgb = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    output_image_rgb = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)

    # Отображаем изображения рядом
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(input_image_rgb)
    plt.title("Вхідне зображення")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(output_image_rgb)
    plt.title("Вихідне зображення")
    plt.axis("off")

    plt.show()
```
